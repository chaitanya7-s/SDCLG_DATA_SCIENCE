```python
#Create a function where a nan value is above 85% then drop column

nan_pr=(df.isnull().sum()/len(df))*100
print(nan_pr)

above_85=nan_pr[nan_pr > 85].index
df=df.drop(columns=above_85)
```

```python
df.isnull().sum()

#Check which row has null values or pick only rows which has null values and after that drop it 
null_value=df.isnull().any(axis=1)
df[null_value]

#If we have more than 2 nan value then we want row
null_value=df.isna().any(axis=1)<2
df=df[null_value]
df.shape

```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.columns
```

```python
x=df.loc[:,['sp500 open', 'sp500 high', 'sp500 low', 'sp500 close',
       'sp500 volume', 'sp500 high-low', 'nasdaq open', 'nasdaq high',
       'nasdaq low', 'nasdaq close', 'nasdaq volume', 'nasdaq high-low',
       'usd_chf', 'eur_usd', 'silver open', 'silver high', 'silver low',
       'silver close', 'silver volume', 'silver high-low', 'oil open',
       'oil high', 'oil low', 'oil close', 'oil volume', 'oil high-low',
       'platinum open', 'platinum high', 'platinum low', 'platinum close',
       'platinum volume', 'platinum high-low', 'palladium open',
       'palladium high', 'palladium low', 'palladium close',
       'palladium volume', 'palladium high-low', 'gold open', 'gold high',
       'gold low', 'gold volume']]
```

```python
y=df['gold close']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test) 
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(LR.score(x_test,y_test))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
